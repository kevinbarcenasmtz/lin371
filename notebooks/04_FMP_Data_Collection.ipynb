```python

def get_earnings_transcript(symbol: str, year: int, quarter: int):
    url = f"https://financialmodelingprep.com/api/v3/earning_call_transcript/{symbol}"
    params = {
        "year": year,
        "quarter": quarter,
        "apikey": api_key
    }
    response = requests.get(url, params=params)
    response.raise_for_status()
    return response.json()

def save_transcript(data: list, quarter: int = None, year: int = None, output_dir: str = "transcripts"):
    os.makedirs(output_dir, exist_ok=True)

    for transcript in data:
        symbol = transcript.get("symbol", "UNKNOWN")
        # Use passed-in quarter/year, fall back to transcript fields
        period = f"Q{quarter}" if quarter else transcript.get("period", "Q?")
        yr = year if year else transcript.get("year", "????")
        content = transcript.get("content", "")

        company_dir = os.path.join(output_dir, f"{symbol}_transcripts")
        os.makedirs(company_dir, exist_ok=True)

        filename = f"{symbol}_{yr}_{period}.txt"
        filepath = os.path.join(company_dir, filename)

        with open(filepath, "w", encoding="utf-8") as f:
            f.write(f"Symbol: {symbol}\n")
            f.write(f"Period: {period} {yr}\n")
            f.write(f"Date: {transcript.get('date', 'N/A')}\n")
            f.write("=" * 60 + "\n\n")
            f.write(content)

        print(f"Saved: {filepath}")

def get_available_transcripts(symbol: str):
    """Get list of all available transcript dates/quarters for a symbol"""
    url = f"https://financialmodelingprep.com/api/v4/earning_call_transcript"
    params = {
        "symbol": symbol,
        "apikey": api_key
    }
    response = requests.get(url, params=params)
    response.raise_for_status()
    return response.json()  # returns list of [year, quarter] pairs

```

```python
# --- Pull ALL Coca-Cola transcripts ---
symbol = "AMZN"

available = get_available_transcripts(symbol)
print(f"Found {len(available)} transcripts for {symbol}")

for entry in available:
    year = entry[1]
    quarter = entry[0]
    print(f"Fetching Q{quarter} {year}...")
    try:
        data = get_earnings_transcript(symbol, year=year, quarter=quarter)
        if data:
            save_transcript(data, quarter=quarter, year=year)  # <-- pass them here
    except Exception as e:
        print(f"  Failed Q{quarter} {year}: {e}")

print("Done!")
```
